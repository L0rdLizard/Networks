## Код номер 1

```python
import numpy as np
from scipy.special import comb

def upper_bound_error_probability(n, p, d):
    upper_bound = 1 - sum(comb(n, i) * (p ** i) * ((1 - p) ** (n - i)) for i in range(d))
    return upper_bound

def exact_error_probability(n, p, d):
    exact_probability = sum(comb(n, i) * (p ** i) * ((1 - p) ** (n - i)) for i in range(d, n + 1))
    return exact_probability

def main():
    n = int(input("Введите длину кодового слова (n): "))
    p = float(input("Введите вероятность ошибки в канале (p): "))
    d = int(input("Введите минимальное расстояние кода (d): "))
    A = [0, 0, 0, 1, 2, 4, 8, 16]  # количество кодовых слов с весом i

    upper_bound = upper_bound_error_probability(n, p, d)

    exact_probability = exact_error_probability(n, p, d)

    print(f"Верхняя оценка вероятности ошибки декодирования: {upper_bound:.6f}")
    print(f"Точное значение вероятности ошибки декодирования: {exact_probability:.6f}")

if __name__ == "__main__":
    main()
```

## Код номер 2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb

def coder(m):
    r = 3
    m = m << r
    c = 1 # TODO
    a = m ^ c
    return a

def calcA(n, l):
    A_arr = []
    for i in range(n+1):
        m = [i for i in range(2**l)]
        m_coded = [coder(i) for i in m]
        temp = 0
        for j in m_coded:
            if bin(j).count('1') == i:
                temp += 1
        print("A{} = {}".format(i, temp))
        A_arr.append(temp)
    return A_arr

def calculate_error_probability(p, d, n, l):
    # A = [1, 0, 0, 7, 7, 0, 0, 1] 
    A = calcA(n, l)
    
    P_e = np.sum([A[i] * (p**i) * ((1-p)**(n-i)) for i in range(d, n+1)])
    
    P_e_upper = np.sum([comb(n, i) * (p**i) * ((1-p)**(n-i)) for i in range(d, n+1)])
    return P_e, P_e_upper

def plot_error_probability(p_values, d, n_values):
    for l in n_values:
        P_e_values = []
        P_e_upper_values = []
        for p in p_values:
            P_e, P_e_upper = calculate_error_probability(p, d, l)
            P_e_values.append(P_e)
            P_e_upper_values.append(P_e_upper)
        plt.plot(p_values, P_e_values, label=f'P_e for l={l}')
        plt.plot(p_values, P_e_upper_values, label=f'P_e_upper for l={l}', linestyle='--')
    plt.xlabel('Вероятность ошибки в канале')
    plt.ylabel('Вероятность ошибки декодирования')
    plt.legend()
    plt.show()

gx = 0b1011    # порождающий многочлен
p_values = np.linspace(0, 1, 100)  # различные значения вероятности ошибки в канале
d = 3    # минимальное расстояние кода
l = 4    # длинна кодируемого сообщения
n_values = [7]  # различные значения длины кодового слова
eps = 0.001

# g(x) = x^3 + x + 1
# d = 3
# k = 4
# n = 7
# r = 3
# l = (2, 3, 4, 5, 6)

# r - степень от g(x)
# k - длинна кодируемого сообщения (l)
# n = k(l) + r 
# n - длина переданного сообщения

plot_error_probability(p_values, d, n_values)
```
